```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 4)
def find_col(df, keywords):
    keywords = [k.lower() for k in keywords]
    for col in df.columns:
        col_low = col.lower().strip()
        for k in keywords:
            if k in col_low:
                return col
    return None

files = {
    2020: "historic_demand_year_2020.csv",
    2021: "historic_demand_year_2021.csv",
    2022: "historic_demand_year_2022.csv",
    2023: "historic_demand_year_2023.csv",
    2024: "historic_demand_year_2024.csv",
}

dfs = []

for year, path in files.items():
    df = pd.read_csv(path)

    date_col   = find_col(df, ["settlement_date", "date"])
    period_col = find_col(df, ["settlement_period", "period"])
    demand_col = find_col(df, ["england_wales_demand", "demand"])

    if not date_col or not period_col or not demand_col:
        raise ValueError(
            f"Erro nas colunas do arquivo {path}.\n"
            f"Colunas encontradas: {list(df.columns)}"
        )

    df[date_col] = pd.to_datetime(df[date_col])

    period = df[period_col] - 1
    df["hour"]   = (period // 2).astype(int)
    df["minute"] = ((period % 2) * 30).astype(int)

    df["datetime"] = (
        df[date_col].dt.normalize()
        + pd.to_timedelta(df["hour"],   unit="h")
        + pd.to_timedelta(df["minute"], unit="m")
    )


    df["demand_MW"]  = df[demand_col]
    df["energy_MWh"] = df["demand_MW"] * 0.5


    df["year"]       = year
    df["month"]      = df["datetime"].dt.month
    df["hour_of_day"] = df["datetime"].dt.hour
    df["weekday"]    = df["datetime"].dt.weekday

    dfs.append(df)


df_all = pd.concat(dfs).sort_values("datetime").reset_index(drop=True)
print(" Dimensão total do conjunto de dados:", df_all.shape)


print("\n==============================")
print("    ENERGIA TOTAL POR ANO")
print("==============================")

annual = df_all.groupby("year")["energy_MWh"].sum().round(0)
display(annual.to_frame("Energia (MWh)"))

annual.plot(kind="bar", color="steelblue")
plt.title("Energia total consumida por ano (MWh)")
plt.ylabel("MWh")
plt.xlabel("Ano")
plt.tight_layout()
plt.show()


print("\n==============================")
print("    ENERGIA MENSAL POR ANO")
print("==============================")

monthly = df_all.groupby(["year", "month"])["energy_MWh"].sum().round(0)
monthly = monthly.unstack(0)
display(monthly)

plt.figure()
for year in monthly.columns:
    plt.plot(monthly.index, monthly[year], marker="o", label=str(year))

plt.title("Consumo mensal (2020–2024)")
plt.xlabel("Mês")
plt.ylabel("MWh")
plt.xticks(range(1, 13))
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


print("\n==============================")
print("    PERFIL DE DEMANDA POR HORA")
print("==============================")

hourly = (
    df_all.groupby(["year", "hour_of_day"])["demand_MW"]
    .mean()
    .round(1)
    .unstack(0)
)
display(hourly)

plt.figure()
for year in hourly.columns:
    plt.plot(hourly.index, hourly[year], marker=".", label=str(year))

plt.title("Perfil médio diário de demanda por ano")
plt.xlabel("Hora do dia")
plt.ylabel("Demanda média (MW)")
plt.xticks(range(0, 24))
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


print("\n==============================")
print("    DIAS ÚTEIS vs FINAIS DE SEMANA")
print("==============================")

df_all["is_weekend"] = df_all["weekday"] >= 5

weekday_p = df_all[~df_all["is_weekend"]].groupby("hour_of_day")["demand_MW"].mean()
weekend_p = df_all[df_all["is_weekend"]].groupby("hour_of_day")["demand_MW"].mean()

plt.figure()
plt.plot(weekday_p.index, weekday_p.values, label="Dias úteis", linewidth=2)
plt.plot(weekend_p.index, weekend_p.values, label="Fins de semana", linewidth=2)
plt.title("Perfil médio: dias úteis x finais de semana")
plt.xlabel("Hora do dia")
plt.ylabel("Demanda média (MW)")
plt.xticks(range(0, 24))
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


print("\n==============================")
print("    TOP 5 PICOS DE DEMANDA POR ANO")
print("==============================")

for year in sorted(df_all["year"].unique()):
    print(f"\n Ano {year}")
    peaks = (
        df_all[df_all["year"] == year]
        .nlargest(5, "demand_MW")[["datetime", "demand_MW"]]
    )
    display(peaks)


print("\n==============================")
print("    SIMULAÇÃO DE AUTOMAÇÃO (15% OFF-HOURS)")
print("==============================")

df_all["is_office_hour"] = df_all["hour_of_day"].between(8, 17)

tarifa   = 600
fator_CO2 = 0.3
reducao  = 0.15

result = []

for year in sorted(df_all["year"].unique()):
    sub = df_all[df_all["year"] == year]

    off_hours_energy = sub[~sub["is_office_hour"]]["energy_MWh"].sum()

    saved_energy = off_hours_energy * reducao
    saved_reais  = saved_energy * tarifa
    saved_CO2    = saved_energy * fator_CO2

    result.append([year, off_hours_energy, saved_energy, saved_reais, saved_CO2])

df_results = pd.DataFrame(
    result,
    columns=["Ano", "Off-hours (MWh)", "Economia MWh", "Economia Reais", "Redução CO₂ (t)"]
).round(0)

display(df_results)

plt.figure()
plt.bar(df_results["Ano"].astype(str), df_results["Economia MWh"], color="orange")
plt.title("Economia estimada de energia por ano (15% off-hours)")
plt.xlabel("Ano")
plt.ylabel("MWh economizados")
plt.tight_layout()
plt.show()

```
